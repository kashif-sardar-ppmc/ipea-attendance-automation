```python
import pandas as pd
import json

df = pd.read_csv("attendance_2026-01-01_to_2026-01-30.csv", header=1)

# Clean headers
df.columns = df.columns.str.strip()
df = df.dropna(how="all").reset_index(drop=True)

employee = {
    "employee_id": df.iloc[0][df.columns[1]],
    "employee_name": df.iloc[0][df.columns[2]],
    "department": df.iloc[0][df.columns[3]],
    "attendance": {}
}

cols = list(df.columns)
i = 4

while i < len(cols) - 1:  # prevent index error

    date = cols[i]

    # Skip if this column doesn't look like a date
    if not str(date).startswith("2026"):
        i += 1
        continue

    in_time = df.iloc[0][cols[i]]

    # Check if OUT column exists
    if i + 1 < len(cols):
        out_time = df.iloc[0][cols[i + 1]]
    else:
        out_time = None

    # Handle absent / empty
    if pd.isna(in_time) or str(in_time).lower() == "absent":
        in_time = "0:00"

    if pd.isna(out_time):
        out_time = "0:00"

    employee["attendance"][date] = {
        "in": str(in_time),
        "out": str(out_time)
    }

    i += 3  # move to next date block

print(json.dumps(employee, indent=2))
```

```python
import pandas as pd
import json

# ==============================
# 1. Read CSV with 2 header rows
# ==============================
file_path = "attendance_2026-01-01_to_2026-01-30.csv"
df = pd.read_csv(file_path, header=[0, 1])

# ==============================
# 2. Flatten multi-level columns
# ==============================
new_cols = []
for col1, col2 in df.columns:
    col1 = str(col1).strip()
    col2 = str(col2).strip()
    if col2 == "" or "Unnamed" in col2:
        new_cols.append(col1)
    else:
        new_cols.append(f"{col1}_{col2}")

df.columns = new_cols
df = df.dropna(how="all").reset_index(drop=True)

# ==============================
# 3. Build JSON for all employees
# ==============================
all_employees = []

for idx, row in df.iterrows():
    employee = {
        "employee_id": row["Employee ID"],
        "employee_name": row["Employee Name"],
        "department": row["Department"],
        "attendance": {}
    }

    for col in df.columns:
        if col.endswith("_IN"):
            date = col.replace("_IN", "")
            in_time = row[col]

            out_col = f"{date}_OUT"
            hrs_col = f"{date}_HRS"

            # Get OUT and HRS if columns exist
            out_time = row[out_col] if out_col in df.columns else None
            hrs = row[hrs_col] if hrs_col in df.columns else None

            # Handle explicit 'Absent'
            if str(in_time).lower() == "absent":
                in_time = "0:00"
                out_time = "0:00"
                hrs = "0:00"
            else:
                # Replace dash or NaN with None
                if pd.isna(in_time) or in_time == "-":
                    in_time = None
                if pd.isna(out_time) or out_time == "-":
                    out_time = None
                if pd.isna(hrs) or hrs == "-":
                    hrs = None

            employee["attendance"][date] = {
                "in": str(in_time) if in_time else None,
                "out": str(out_time) if out_time else None,
                "hrs": str(hrs) if hrs else None
            }

    all_employees.append(employee)

# ==============================
# 4. Save JSON
# ==============================
with open("attendance.json", "w") as f:
    json.dump(all_employees, f, indent=2)

# Print first employee for verification
print(json.dumps(all_employees[0], indent=2))
```

```python
import csv
import json

def transform_attendance(file_path):
    with open(file_path, mode='r', encoding='utf-8') as f:
        reader = list(csv.reader(f))
        
        # Row 0: Dates (e.g., Sr, Employee ID, ..., 2026-01-01, , , 2026-01-02)
        # Row 1: Labels (e.g., , , , , IN, OUT, HRS, IN, OUT, HRS)
        header_dates = reader[0]
        data_rows = reader[2:] # Actual data starts from index 2
        
        employees_json = []

        for row in data_rows:
            # Skip empty rows or footer rows (like the PPMC Synergy download note)
            if not row or len(row) < 4 or not row[1].isdigit():
                continue
                
            emp_data = {
                "employee_id": float(row[1]),
                "employee_name": row[2],
                "department": row[3],
                "attendance": {}
            }
            
            # Attendance data starts at index 4
            # We jump 3 columns at a time (IN, OUT, HRS)
            col_idx = 4
            for i in range(4, len(header_dates) - 1, 3):
                date_label = header_dates[i].strip()
                if not date_label:
                    continue
                
                # Get the triplet
                val_in = row[col_idx] if col_idx < len(row) else "-"
                val_out = row[col_idx + 1] if col_idx + 1 < len(row) else "-"
                val_hrs = row[col_idx + 2] if col_idx + 2 < len(row) else "-"
                
                # Clean "Absent" or "-" values to match your requested format
                if val_in == "Absent":
                    entry = {"in": "0:00", "out": "0:00", "hrs": "0:00"}
                else:
                    entry = {
                        "in": val_in if val_in != "-" else None,
                        "out": val_out if val_out != "-" else None,
                        "hrs": val_hrs if val_hrs != "-" else None
                    }
                
                emp_data["attendance"][date_label] = entry
                col_idx += 3
            
            employees_json.append(emp_data)
            
        return employees_json

# Execution
try:
    result = transform_attendance('attendance_2026-01-01_to_2026-01-30.csv')
    # Print the first employee record as a sample
    print(json.dumps(result[0], indent=2))
except FileNotFoundError:
    print("Error: file.csv not found.")
```
